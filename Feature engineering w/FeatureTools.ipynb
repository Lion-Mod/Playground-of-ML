```
import featuretools as ft
import pandas as pd
```

```
hr_data = pd.read_csv("hr_data.csv")
```

```
hr_data.head()
```

```
es = ft.EntitySet()
```

```
es = es.entity_from_dataframe(entity_id = "employees",
                              dataframe = hr_data,
                              index = "EmployeeNumber",
                              variable_types={"Age" : ft.variable_types.Numeric,
                                              "Attrition" : ft.variable_types.Boolean,
                                              "DailyRate" : ft.variable_types.Numeric,
                                              "Department" : ft.variable_types.Categorical,
                                              "DistanceFromHome" : ft.variable_types.Numeric,
                                              "Education" : ft.variable_types.Categorical,
                                              "EducationField" : ft.variable_types.Categorical,
                                              "EmployeeCount" : ft.variable_types.Numeric,
                                              "EnvironmentSatisfaction" : ft.variable_types.Categorical,
                                              "Gender" : ft.variable_types.Boolean,
                                              "HourlyRate" : ft.variable_types.Numeric,
                                              "JobInvolvement" : ft.variable_types.Categorical,
                                              "JobLevel" : ft.variable_types.Ordinal,
                                              "JobRole" : ft.variable_types.Categorical,
                                              "JobSatisfaction" : ft.variable_types.Ordinal,
                                              "MaritalStatus" : ft.variable_types.Categorical,
                                              "MonthlyIncome" : ft.variable_types.Numeric,
                                              "MonthlyRate" : ft.variable_types.Numeric,
                                              "NumCompaniesWorked" : ft.variable_types.Numeric,
                                              "Over18" : ft.variable_types.Boolean,
                                              "OverTime" : ft.variable_types.Boolean,
                                              "PercentSalaryHike" : ft.variable_types.Numeric,
                                              "PerformanceRating" : ft.variable_types.Categorical,
                                              "RelationshipSatisfaction" : ft.variable_types.Categorical,
                                              "StandardHours" : ft.variable_types.Numeric,
                                              "StockOptionLevel" : ft.variable_types.Categorical,
                                              "TotalWorkingYears" : ft.variable_types.Numeric,
                                              "TrainingTimesLastYear" : ft.variable_types.Numeric,
                                              "WorkLifeBalance" :	ft.variable_types.Categorical,
                                              "YearsAtCompany" : ft.variable_types.Numeric,
                                              "YearsInCurrentRole" : ft.variable_types.Numeric,
                                              "YearsSinceLastPromotion" : ft.variable_types.Numeric,
                                              "YearsWithCurrManager" : ft.variable_types.Numeric})
```

```
es
```

```
feature_matrix_customers, features_defs = ft.dfs(entityset = es, target_entity = "employees")
```

```
encoded_data = ft.encode_features(feature_matrix_customers, features_defs)
```
